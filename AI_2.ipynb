```python
#jesu li vrijednosti iznad granice detekcije LOD=0,01
def above_LOD(x):
  if x > 0.01:
    return "above lod"
  else:
    return "xxx"

df.applymap(above_LOD)

#df.applymap(above_LOD, inPlace=True) -pamti nove podatke u stari df
```

```python
#stvoriti novi stupac u df ako je above LOD na temelju stupca x
#ako je veći od 0.01 nek piše ta vrijednost ako ne nek pise 0.0001
df["aboveLOD"] = df["x"].apply(lambda x: x if x > 0.01 else "0.0001")
df
```

```python
#grafički prikaz log10 svi vrijednosti osim zadnjeg supcaplt.imshow(np.log10(abs(df.iloc[:,:-1])))
```

```python
#promijenimo stupac aboveLOD - ako je veća 0.01
#mijenjamo postojeći stupac
df["aboveLOD"] = df["x"].apply(lambda x: x if x > 0.01 else "<LOD")
df
```

```python
#novi stupac ako je w veći ili jednak 0 nek piše p ako je manji n
df["positive_nmb"] = df["w"].apply(lambda x: "P" if x >= 0 else "N")
df
```
