```python
from sklearn.impute import SimpleImputer #module importamo iz sklearna
imputer = SimpleImputer(missing_values=np.nan,
                        strategy="median") #naš recept kako popuniti prazne vrijednosti
imputer.fit(X[["Age", "Salary"]]) #fit
imputer.statistics_ # ovdje je naučio medijane naših stupaca prije zamjene i ispisao nam medijane
```

```python
X[["Age", "Salary"]].median() #uglata zagrada u indeksiranju uglata kao pišem listu
```

```python
#fit - učenje
#transform -
X_imputed = imputer.transform(X[["Age", "Salary"]])
X_imputed
```

```python
#BITNO - konverzija matrice u df - data, columns, indexi
X_imp = pd.DataFrame(X_imputed,
                     columns = ["Age", "Salary"],
                     index=X.index)
X_imp
```

```python
#prvo sam prebacio nan u -999
imputer = SimpleImputer(missing_values=np.nan,
                        strategy="constant",fill_value= -999)
imputer.fit(X[["Age", "Salary"]]) #fit
nn = imputer.transform(X[["Age", "Salary"]])
nnn = pd.DataFrame(nn,
                     columns = ["Age", "Salary"],
                     index=X.index)
imputer = SimpleImputer(missing_values=-999,
                        strategy="median")
imputer.fit(nnn[["Age", "Salary"]]) #fit
nn = imputer.transform(nnn[["Age", "Salary"]])
nn
```
