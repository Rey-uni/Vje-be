```python
def chech_gender_by_name(name):
  if name[-1]=="a":
    gender = "female"
  elif name[-1]=="o":
    gender = "male"
  else:
    gender = "unknown"
  return gender
```

```python
def ph(c):
  return -m.log10(c)
```

```python
ph(10)
```

```python
def ph_evaluator(c=0.001):
  ph_value = ph(c)
  if ph_value < 7:
    decision = "kiselo"
  else:
    decision = "lužnato"
  return decision

ph_evaluator(0.1)
```

```python
def ph_evaluator(c=0.001):

  if ph(c) < 7: #zovemo postojeću funkciju:
    return "kiselo"
  else:
    return "lužnato"

conc_list = [0.001, 0.01, 0.1, 1, 10]
for solution in conc_list:
  print(ph_evaluator(solution))
```
